```python
import tensorflow as tf

# Load a sample dataset (e.g., MNIST)
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# Flatten the images
X_train = X_train.reshape(-1, 28*28)
X_test = X_test.reshape(-1, 28*28)

# Normalize the pixel values
X_train = X_train / 255.0
X_test = X_test / 255.0

# Define the input dimension
input_dim = X_train.shape[1]

# Quantum-inspired activation function
def quantum_inspired_activation(x):
  return tf.sin(x) + tf.cos(x)

# Create the model
model = tf.keras.Sequential([
  tf.keras.layers.Dense(64, activation='relu', input_shape=(input_dim,)),
  tf.keras.layers.Dense(32, activation=quantum_inspired_activation),
  tf.keras.layers.Dense(10, activation='softmax')  # 10 classes for MNIST
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print("Test Accuracy:", test_acc)
```
